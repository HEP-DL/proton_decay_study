```python
import pandas
import json
import glob
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context('talk')
sns.set_style('whitegrid')
%matplotlib notebook
```

```python
fig = plt.figure()
ax = plt.subplot()
dat={}
with open('/Users/wier702/Desktop/test_gen.csv','r') as input:
    dat['times'] = [float(i) for i in input.readlines()]
    
df = pandas.DataFrame(dat)
labels=["Marianas Delay"]
df.hist('times', ax=ax)
lines, _ = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')  # legend for first two lines only 
plt.title("File IO Delay")
plt.xlabel("Time [s]")
plt.tight_layout()
```

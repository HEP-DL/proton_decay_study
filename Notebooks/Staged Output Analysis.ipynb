```python
import pandas
import glob
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_context('notebook')
sns.set_style('whitegrid')
%matplotlib notebook
```

```python
files = glob.glob("../../tmp/stage*.csv")
```

```python
fig = plt.figure()
ax = plt.subplot()
labels=[]
for _f in files:
    stage= int(_f.split('stage')[1].strip('.csv'))
    labels.append("Stage {}".format(stage))
    df = pandas.read_csv(_f)
    df.plot('epoch','acc', ax=ax)
lines, _ = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')  # legend for first two lines only 
```

```python
fig = plt.figure()
ax = plt.subplot()
labels=[]
for _f in files:
    stage= int(_f.split('stage')[1].strip('.csv'))
    labels.append("Stage {}".format(stage))
    df = pandas.read_csv(_f)
    df.plot('epoch','loss', ax=ax)
lines, _ = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')  # legend for first two lines only 
```

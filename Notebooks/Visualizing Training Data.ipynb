```python
import pandas
import glob
import seaborn as sns
import os
import matplotlib.pyplot as plt

sns.set_context('poster')
sns.set_style('whitegrid')
%matplotlib notebook
```

```python
basedir = '/Users/wier702/Downloads/'
```

```python
acc=[]
loss=[]
for filename in glob.glob(os.path.join(basedir, "*.csv")):
    print(filename)
    df = pandas.read_csv(filename)
    if 'acc' in filename:
        acc.append(df['Value'])
    else:
        loss.append(df['Value'])
```

```python
ordered_acc=sorted(acc, key= lambda x: sum(x)/len(x))
ordered_loss=sorted(loss, key= lambda x: sum(x)/len(x), reverse=True)
```

```python
current_iter = 0
plt.figure()
ax=plt.subplot()
stage=0
for i in ordered_acc:
    x=[(current_iter+it) for it, value in enumerate(i) ]
    y = [j for j in i]
    plt.plot(x,y, label="Stage {}".format(stage))
    stage=stage+1
    current_iter=current_iter+len(i)
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
ax.set_yticklabels([])
ax.set_xticklabels([])
plt.tight_layout()
```

```python
current_iter = 0
plt.figure()
ax=plt.subplot()
stage=0
for i in ordered_loss:
    x=[(current_iter+it) for it, value in enumerate(i) ]
    y = [j for j in i]
    y = [j for j in reversed(y)]
    plt.plot(x,y, label="Stage {}".format(stage))
    stage=stage+1
    current_iter=current_iter+len(i)
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Loss")
ax.set_yticklabels([])
ax.set_xticklabels([])
plt.tight_layout()
```
